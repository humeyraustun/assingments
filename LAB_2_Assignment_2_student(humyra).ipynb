## 2: Plot 12 most dangerous cities

```python
#kill.city.value_counts()[:12]
dngrcity= kill.city.value_counts().head(12)
dngrcity
```

```python
# with seaborn countplot
fig, ax = plt.subplots(figsize=(9,5))
sns.countplot( x = "city", data = kill, order = dngrcity.index)
ax.set_xticklabels(dngrcity.index, rotation = 45)
ax.set_title ( "Most Dangereous City", color = "b", fontsize = 15)
```

## 3: Plot 20 most dangerous states

```python
dngrstate= kill.state. value_counts().head(20)
dngrstate
```

```python
# with seaborn barplot
fig, ax = plt.subplots(figsize=(9,5))
sns.barplot(x = dngrstate.index, y = dngrstate.values)
ax.set_xticklabels(dngrstate.index, rotation = 45)
ax.set_title ( "Most Dangereous states", color = "b", fontsize = 15)
plt.show()
```

```python
# with seaborn countplot
fig, ax = plt.subplots(figsize=(9,5))
sns.countplot( x = "state", data = kill, order = dngrstate.index)
ax.set_xticklabels(dngrstate.index, rotation = 45)
ax.set_title ( "Most Dangereous state", color = "b", fontsize = 15)
plt.show()
```
